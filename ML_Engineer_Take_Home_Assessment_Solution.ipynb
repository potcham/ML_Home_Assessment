```python
import os
import torch
import torch.nn as nn
import numpy as np
import torchvision
from torchvision.models import resnet18, ResNet18_Weights
```

```python
class DetectorModel(nn.Module):

  def __init__(self, backbone=None, num_classes=10):
    super(DetectorModel, self).__init__()
    self.backbone = backbone
    self.num_classes = num_classes

    # Regresor Head
    self.regressor = nn.Sequential(
                      nn.Linear(self.backbone.fc.in_features, 128),
                      nn.ReLU(),
                      nn.Linear(128, 64),
                      nn.ReLU(),
                      nn.Linear(64, 32),
                      nn.ReLU(),
                      nn.Linear(32, 4),
                      nn.Sigmoid()
		)
    # Classification Head
    self.classifier = nn.Sequential(
                      nn.Linear(self.backbone.fc.in_features, 512),
                      nn.ReLU(),
                      nn.Dropout(),
                      nn.Linear(512, 512),
                      nn.ReLU(),
                      nn.Dropout(),
                      nn.Linear(512, self.num_classes)
		)

    self.backbone.fc = nn.Identity()

  def forward(self, x):
    features = self.backbone(x)
    b_boxes = self.regressor(features)
    class_logits = self.classifier(features)

    return (b_boxes, class_logits)
```

### Custom Dataset class

```python
from PIL import Image

import torchvision.transforms as transforms
from torch.utils.data import Dataset

class CustomDataset(Dataset):

  def __init__(self, dataset_path, images_list, transforms=None, class_idx=None):
    self.image_root = os.path.join(dataset_path, "images")
    self.label_root = os.path.join(dataset_path, "labels")
    self.images_list = images_list
    self.transforms = transforms
    self.class_to_idx = class_idx

  def read_xml_file(self, path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }

  def __len__(self):
    return len(self.images_list)

  def __getitem__(self, index):
    img_path = os.path.join(self.image_root, self.images_list[index]+'.jpg')
    image = Image.open(img_path).convert("RGB")
    W, H = image.size

    annot_path = os.path.join(self.label_root, self.images_list[index]+'.xml')
    annot = self.read_xml_file(annot_path)

    boxes = (annot['xmin']/W, annot['ymin']/H, annot['xmax']/W, annot['ymax']/H)
    boxes = torch.as_tensor(boxes, dtype=torch.float32)

    label = self.class_to_idx[annot['specie']]
    labels = torch.zeros((len(self.class_to_idx),), dtype=torch.float32)
    labels[label] = 1

    if self.transforms:
      image = self.transforms(image)
    else:
      image = transforms.PILToTensor(image)

    return (image, labels, boxes)
```

```python
weights = torchvision.models.ResNet18_Weights.DEFAULT
base_model = torchvision.models.resnet18(weights=weights)

for param in base_model.parameters():
  param.requires_grad = False

model = DetectorModel(base_model, num_classes=len(class_names)).to(device)
```

```python
LR = 1e-4
EPOCHS = 10
CLS_LOSS_W = 1
BBOX_LOSS_W = 1

class_loss_fn = nn.CrossEntropyLoss()
bbox_loss_fn = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

```python
from typing import Tuple, List, Dict
from tqdm.auto import tqdm

def train_step(model: torch.nn.Module, 
               dataloader: torch.utils.data.DataLoader, 
               loss_fn: List[torch.nn.Module], 
               optimizer: torch.optim.Optimizer,
               device: torch.device) -> Tuple[float, float]:

    model.train()

    train_loss, train_acc = 0, 0

    for batch, (X, y, bboxes) in enumerate(dataloader): #(image, labels, boxes)
        # Send data to target device
        X, y, bboxes = X.to(device), y.to(device), bboxes.to(device)

        # 1. Forward pass
        bboxes_pred, cls_pred = model(X)

        # 2. Calculate  and accumulate loss
        # print(bboxes_pred.size(), bboxes.size(), bboxes_pred.dtype, bboxes.dtype)
        # print(cls_pred.size(), y.size(), cls_pred.dtype, y.dtype)
        bbox_loss = loss_fn[0](bboxes_pred, bboxes)
        cls_loss = loss_fn[1](cls_pred, y)
        
        total_loss = bbox_loss*BBOX_LOSS_W + cls_loss*CLS_LOSS_W
        train_loss += total_loss.item() 

        # 3. Optimizer zero grad
        optimizer.zero_grad()

        # 4. Loss backward
        total_loss.backward()

        # 5. Optimizer step
        optimizer.step()

        # Calculate and accumulate accuracy metric across all batches
        y_pred_class = torch.argmax(torch.softmax(cls_pred, dim=1), dim=1)
        train_acc += (y_pred_class == y.argmax(dim=1)).sum().item()/len(cls_pred)

    # Adjust metrics to get average loss and accuracy per batch 
    train_loss = train_loss / len(dataloader)
    train_acc = train_acc / len(dataloader)
    return train_loss, train_acc

def test_step(model: torch.nn.Module, 
              dataloader: torch.utils.data.DataLoader, 
              loss_fn: List[torch.nn.Module],
              device: torch.device) -> Tuple[float, float]:

    model.eval() 

    test_loss, test_acc = 0, 0

    # Turn on inference context manager
    with torch.inference_mode():
        # Loop through DataLoader batches
        for batch, (X, y, bboxes) in enumerate(dataloader):
            # Send data to target device
            X, y, bboxes = X.to(device), y.to(device), bboxes.to(device)

            # 1. Forward pass
            bboxes_pred, cls_pred_logits = model(X)

            # 2. Calculate and accumulate loss
            bbox_loss = loss_fn[0](bboxes_pred, bboxes)
            cls_loss = loss_fn[1](cls_pred_logits, y)
            
            total_loss = bbox_loss*BBOX_LOSS_W + cls_loss*CLS_LOSS_W
            test_loss += total_loss.item()

            # Calculate and accumulate accuracy
            test_pred_labels = cls_pred_logits.argmax(dim=1)
            test_acc += ((test_pred_labels == y.argmax(dim=1)).sum().item()/len(test_pred_labels))

    # Adjust metrics to get average loss and accuracy per batch 
    test_loss = test_loss / len(dataloader)
    test_acc = test_acc / len(dataloader)
    return test_loss, test_acc

def train(model: torch.nn.Module, 
          train_dataloader: torch.utils.data.DataLoader, 
          test_dataloader: torch.utils.data.DataLoader, 
          optimizer: torch.optim.Optimizer,
          loss_fn: List[torch.nn.Module],
          epochs: int,
          device: torch.device) -> Dict[str, List]:

    # Create empty results dictionary
    results = {"train_loss": [],
      "train_acc": [],
      "test_loss": [],
      "test_acc": []
    }

    # Loop through training and testing steps for a number of epochs
    for epoch in tqdm(range(epochs)):
        train_loss, train_acc = train_step(model=model,
                                          dataloader=train_dataloader,
                                          loss_fn=loss_fn,
                                          optimizer=optimizer,
                                          device=device)
        test_loss, test_acc = test_step(model=model,
          dataloader=test_dataloader,
          loss_fn=loss_fn,
          device=device)

        # Print out what's happening
        print(
          f"Epoch: {epoch+1} | "
          f"train_loss: {train_loss:.4f} | "
          f"train_acc: {train_acc:.4f} | "
          f"test_loss: {test_loss:.4f} | "
          f"test_acc: {test_acc:.4f}"
        )

        # Update results dictionary
        results["train_loss"].append(train_loss)
        results["train_acc"].append(train_acc)
        results["test_loss"].append(test_loss)
        results["test_acc"].append(test_acc)

    # Return the filled results at the end of the epochs
    return results
```

```python
import cv2

def pred_and_plot_image(model: torch.nn.Module,
                        image_path: str, 
                        image_size: Tuple[int, int] = (224, 224),
                        conf: float = 0.5,
                        transform: torchvision.transforms = None,
                        device: torch.device=device):
    
    if transform is not None:
        image_transform = transform
    else:
        image_transform = transforms.Compose([
            transforms.Resize(image_size),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225]),
        ])
    
    # image_path = os.path.join('assessment_dataset','images','pomeranian_18.jpg')
    img = Image.open(image_path)
    W,H = img.size

    model.eval()
    with torch.inference_mode():
      # 6. Transform and add an extra dimension to image (model requires samples in [batch_size, color_channels, height, width])
      transformed_image = transform(img).unsqueeze(dim=0)
      bboxes_pred, cls_pred_logits = model(transformed_image.to(device))

    target_image_pred_probs = torch.softmax(cls_pred_logits, dim=1)
    target_image_pred_label = torch.argmax(target_image_pred_probs, dim=1)
    prob = target_image_pred_probs.max().cpu().item()
    label = target_image_pred_label.cpu().numpy()[0]

    bboxes = bboxes_pred.cpu().numpy()[0]
    x_min = int(bboxes[0]*W)
    y_min = int(bboxes[1]*H)
    x_max = int(bboxes[2]*W)
    y_max = int(bboxes[3]*H)

    output = { "has_object": False, "cat_or_dog": "", "specie": "", "xmin": 0, "ymin": 0, "xmax": 0, "ymax": 0 }
    img = cv2.imread(image_path)

    if prob >= conf:
      cv2.rectangle(img, (x_min,y_min),(x_max,y_max), (255,0,0), 2)

      specie = idx_to_class[label]
      cat_dog = 'cat' if specie in cat_list else 'dog'

      output = { "has_object": True, "cat_or_dog": cat_dog, "specie": specie, "xmin": x_min, "ymin": y_min, "xmax": x_max, "ymax": y_max }

    plt.imshow(img[:,:,::-1])

    return output
```
